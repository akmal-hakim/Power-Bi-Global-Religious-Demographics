```python
import pandas as pd
df = pd.read_csv('GlobalReligion.csv')
df.head()
```

```python
print('Number of rows and column :')
print({df.shape[0], df.shape[1]})
print('Number of duplicate entries in the dataset are : ')
print(df.duplicated().sum())
print('Number missing values in the dataset are : ')
print(sum(df.isna().sum()))
```

```python
missing_val = df.isnull().sum()[df.isnull().sum() > 0]
print('missing values by feature')
print(missing_val)
```

```python
df.fillna(0, inplace = True)
missing_val = df.isnull().sum()[df.isnull().sum() > 0]
print(missing_val)
```

```python
print('Number of rows and column :')
print({df.shape[0], df.shape[1]})
print('Number of duplicate entries in the dataset are : ')
print(df.duplicated().sum())
print('Number missing values in the dataset are : ')
print(sum(df.isna().sum()))
```

```python
df.to_csv('Cleaned_Global_Religion.csv')
```
